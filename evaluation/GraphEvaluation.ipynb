# Weird places

```python
gdf_results = gpd.read_parquet("results/hexagon_grid_bamberg.parquet")
gdf_results.head(1)
```

```python
gdf_266 = gdf_results[gdf_results.index == 266]
m = gdf_266.explore()
gdf_266.centroid.explore(m=m, color="red", style_kwds={"radius":4})
```

```python
with open("results/walk.p", "rb") as file:
    G_walk = pickle.load(file)
```

```python
gdf_nodes = ox.graph_to_gdfs(G_walk, edges=False).to_crs(gdf_266.crs)
```

```python
gdf_nodes[gdf_nodes.distance(gdf_266.centroid.iloc[0]) < 100].explore(m=m, color="blue")
```

```python
gdf_edges = ox.graph_to_gdfs(G_walk, nodes=False).to_crs(gdf_266.crs)
```

```python
gdf_edges[gdf_edges.distance(gdf_266.centroid.iloc[0]) < 100].explore(m=m, color="green")
```

```python
gdf_pois = gpd.read_parquet("data/osm_features/bamberg_osm_features.parquet").to_crs(gdf_266.crs)
```

```python
gdf_pois[["category", "geometry"]][(gdf_pois["category"] == "education") & (gdf_pois.distance(gdf_266.centroid.iloc[0]) < 1000)].explore(m=m, color="purple")
```
