```python
import geopandas as gpd
import shapely
```

```python
gdf_buildings = gpd.GeoDataFrame(geometry=[shapely.Polygon([(0,3), (1,3), (1,2), (0,2)]), 
                                 shapely.Polygon([(1,3), (2,3), (2,2.25), (1,2.25)]),
                                 shapely.Polygon([(2,3), (3,3), (3, 1.75), (2,2)]),
                                 shapely.Polygon([(3,3), (4,3), (4, 2), (3, 2)]), # up to here upper row
                                 shapely.Polygon([(0,0), (0.5,0), (0.5,1), (0,1)]), 
                                 shapely.Polygon([(0.5,0), (1.75,0), (1.75,0.75), (0.5,0.75)]),
                                 shapely.Polygon([(1.75,0), (3,0), (3, 0.75), (1.75,0.75)]),
                                 shapely.Polygon([(3,0), (4,0), (4, 1), (3, 1)]),
                                 ])

gdf_buildings.plot(cmap="Pastel1")
```

```python
gdf_street = gpd.GeoDataFrame(geometry=[shapely.LineString([(0, 1.5), (4, 1.5)])])
g = gdf_street.plot(color="black")
gdf_buildings.plot(ax=g, cmap="Pastel1")
```

```python
gdf_street["geometry"] = gdf_street.geometry.buffer(0.5)
g = gdf_street.plot(color="black")
gdf_buildings.plot(ax=g, cmap="Pastel1")
```

```python
# gdf_street = gdf_street["geometry"].clip(gdf_buildings.reverse())
gdf_street["geometry"] = gdf_street.geometry.difference(gdf_buildings.union_all())
g = gdf_street.plot(color="black")
gdf_buildings.plot(ax=g, cmap="Pastel1")
gdf_street.plot(color="black")
```
